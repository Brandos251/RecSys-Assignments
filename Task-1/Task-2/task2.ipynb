```python

matrix = [[0, 0.88, 0.93, 0.87, 0.87],
          [0, 0, 0.88, 0.85, 0.86],
          [0, 0, 0, 0.88, 0.86],
          [0, 0, 0, 0, 0.83],
          [0, 0, 0, 0, 0]]

users = {0: 'U1', 1: 'U2', 2: 'U3', 3: 'U4', 4: 'U5'}

#Размер кластера
r = 0.85

union_users = ['U1']

# Функция для поиска максимального значения в матрице и его индекса.
def find_max_and_position(mat):
    max_val = -1  
    position = [-1, -1] 
    
    
    for row_index in range(len(mat)):
        for col_index in range(len(mat[row_index])):
            current = mat[row_index][col_index]
            # Если текущий элемент больше максимального, обновляем максимум и позицию
            if current > max_val:
                max_val = current
                position[0], position[1] = row_index, col_index
    
    return max_val, position

# Функция для удаления пользователя из матрицы.
def remove_user_from_matrix(mat, user_index):
    updated_matrix = []  # Матрица после удаления
    
    
    for row_idx in range(len(mat)):
        if row_idx == user_index[1]:
            continue
        
        new_row = []  # Новая строка после удаления столбца
        for col_idx in range(len(mat[row_idx])):
            if col_idx != user_index[1]:
                new_row.append(mat[row_idx][col_idx])
        
        # Добавляем строку в обновленную матрицу
        if new_row:
            updated_matrix.append(new_row)
    
    return updated_matrix

# Функция для перестройки словаря пользователей
def rebuild_users_dict(users, removed_index):
    new_users = {}  
    counter = 0 
    
    # Проходим по всем пользователям и добавляем только тех, кого не удаляем
    for idx in users:
        if idx != removed_index:
            new_users[counter] = users[idx]  
            counter += 1  
    
    return new_users

# Функция для отображения матрицы
def display_matrix(mat):
    for row in mat:
        formatted_row = ''.join([str(val).ljust(6) for val in row])
        print(formatted_row)


def process_matrix(matrix, users, threshold):
    iteration_count = 1  
    
    while True:
        # Находим максимум в матрице и проверяем, превышает ли он пороговое значение
        max_value, max_position = find_max_and_position(matrix)
        if max_value >= threshold:
        
            matrix = remove_user_from_matrix(matrix, max_position)
            
            print(f'Итерация {iteration_count}. Перестройка матрицы расстояний:')
            print(f'{union_users[-1]} и {users[max_position[1]]} объединены по расстоянию {max_value} >= R = {threshold}')
            
            iteration_count += 1
            union_users.append(f'({union_users[-1]}, {users[max_position[1]]})')
            
            # Перестраиваем словарь пользователей
            users = rebuild_users_dict(users, max_position[1])
            
            display_matrix(matrix)
            print('-' * 50)
        else:
            print('-' * 50, 'Завершено', sep='')
            break
```
